```python

def read_images(directory):
    images = []
    for filename in os.listdir(directory):
        if filename.endswith(('.png', '.jpg', '.jpeg')):
            image_path = os.path.join(directory, filename)
            img = Image.open(image_path)
            images.append(img)
    return images

def display_images(images, titles, rows, cols):
    fig = plt.figure(figsize=(12, 8))
    gs = gridspec.GridSpec(rows, cols, wspace=0.1, hspace=0.4)

    for i in range(rows):
        for j in range(cols):
            ax = plt.subplot(gs[i, j])
            ax.imshow(images[i * cols + j])
            ax.set_title(titles[i * cols + j])
            ax.axis('off')

    plt.savefig("../../SNPimagesExample.pdf", bbox_inches='tight')  # Save as PDF
    plt.show()

if __name__ == "__main__":
    image_directory = "../../SNPimg/dataset/NSFTV_1/"
    images = read_images(image_directory)
    image_titles = ["Chromosome 1", "Chromosome 2", "Chromosome 3", "Chromosome 4",
                    "Chromosome 5", "Chromosome 6", "Chromosome 7", "Chromosome 8",
                    "Chromosome 9", "Chromosome 10", "Chromosome 11", "Chromosome 12"]

    rows = 3
    cols = 4

    display_images(images, image_titles, rows, cols)
```
